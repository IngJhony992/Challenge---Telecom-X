#ðŸ“Œ ExtracciÃ³n

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

datos = pd.read_json('TelecomX_Data.json')
datos
```

#ðŸ”§ TransformaciÃ³n

```python
datos.info()
```

## NormalizaciÃ³n de datos

```python
datos_normalizados_customer = pd.json_normalize(datos['customer'])
datos_normalizados_customer.head()
```

```python
datos_normalizados_customer.info()
```

```python
datos_normalizados_phone = pd.json_normalize(datos['phone'])
datos_normalizados_phone.head()
```

```python
datos_normalizados_phone.info()
```

```python
datos_normalizados_internet = pd.json_normalize(datos['internet'])
datos_normalizados_internet.head()
```

```python
datos_normalizados_internet.info()
```

```python
datos_normalizados_account = pd.json_normalize(datos['account'])
datos_normalizados_account.head()
```

```python
datos_1 = datos.iloc[:,0:2]
datos_1
```

```python
datos_normalizados = pd.concat([datos_1, datos_normalizados_customer, datos_normalizados_phone, datos_normalizados_internet, datos_normalizados_account], axis=1)
datos_normalizados
```

```python
datos_normalizados.columns = datos_normalizados.columns.str.lower().str.replace('.', '_')
```

```python
print(datos_normalizados.columns)
```

```python
datos_normalizados = datos_normalizados.rename(columns={
    'customerid': 'customer_id',
    'phoneservice': 'phone_service',
    'seniorcitizen': 'senior_citizen',
    'multiplelines': 'multiple_lines',
    'internetservice': 'internet_service',
    'onlinesecurity': 'online_security',
    'onlinebackup': 'online_backup',
    'deviceprotection': 'device_protection',
    'techsupport': 'tech_support',
    'paymentmethod': 'payment_method',
    'streamingtv': 'streaming_tv',
    'streamingmovies': 'streaming_movies',
    'paperlessbilling': 'paperless_billing',
})
datos_normalizados.head()
```

```python
datos_normalizados['daily_accounts'] = datos_normalizados['charges_monthly'] / 30
datos_normalizados['daily_accounts'] = datos_normalizados['daily_accounts'].round(2)
datos_normalizados.head(1)
```

```python
datos_normalizados['charges_total'] = datos_normalizados['charges_total'].str.strip().replace('', np.nan).astype(float)
datos_normalizados.info()
```

```python
datos_normalizados = datos_normalizados[datos_normalizados['churn'].astype(str).str.strip() != '']
datos_normalizados.head(5)
```

```python
datos_normalizados.info()
```

#ðŸ“Š Carga y anÃ¡lisis

### AnÃ¡lisis Descriptivo

```python
datos_normalizados.describe()
```

### DistribuciÃ³n de evasiÃ³n

```python
evasion = datos_normalizados['churn'].value_counts().reset_index()
evasion.columns = ['evasion', 'cantidad']
evasion
```

```python
total = evasion['cantidad'].sum()
evasion['porcentaje'] = ((evasion['cantidad'] / total) * 100).round(2)
evasion
```

```python
sizes = evasion['porcentaje']
etiquetas = ['Usuarios que continuan con el servicio', 'Usuarios que no continuan con el servicio']
colores =['green', 'silver']

fig, ax = plt.subplots(figsize = (10,6))
fig.patch.set_facecolor('white')

ax.pie(sizes, labels=etiquetas, autopct='%1.1f%%', colors=colores, textprops={'fontsize':12})
ax.set_title('Porcentaje de distribuciÃ³n de evasiÃ³n de usuarios',
             loc='center', fontsize=18)

plt.show()
```

### Recuento de evasiÃ³n por variables categÃ³ricas

```python
datos_a_dinamizar = datos_normalizados.iloc[:,[1,2,16,18,19]]
datos_a_dinamizar
```

```python
datos_a_dinamizar.head(1)
```

```python
datos_a_dinamizar['churn'] = datos_a_dinamizar['churn'].replace({'Yes': 'SÃ­'})
datos_a_dinamizar['gender'] = datos_a_dinamizar['gender'].replace({'Male': 'Masculino', 'Female': 'Femenino'})
datos_a_dinamizar['contract'] = datos_a_dinamizar['contract'].replace({'Month-to-month': 'Mensual', 'One year': 'Anual', 'Two year': 'Dos aÃ±os'})
datos_a_dinamizar
```

```python
evasion_variables_categoricas = datos_a_dinamizar.pivot_table(index = 'churn', columns = ['gender','contract'], values = 'charges_monthly')
evasion_variables_categoricas
```

```python
colores = ['#4CAF50', '#F44336']

fig, axs = plt.subplots(1, 2, figsize=(12,6), sharey=True)

for i, gender in enumerate(evasion_variables_categoricas.columns.levels[0]):
    subtabla = evasion_variables_categoricas[gender]
    subtabla.T.plot(kind='bar', ax=axs[i], color=colores)

    axs[i].set_title(f'GÃ©nero: {gender}')
    axs[i].set_xlabel('Tipo de contrato', fontsize=12)
    axs[i].set_ylabel('Cargo mensual promedio', fontsize=12)
    axs[i].legend(title='EvasiÃ³n')
    axs[i].tick_params(axis='x', rotation=0)

    for rect in axs[i].patches:
        height = rect.get_height()
        x = rect.get_x() + rect.get_width() / 2

        axs[i].text(x,
            height + 1,
            f'{height:.1f}',
            ha='center',
            va='bottom',
            fontsize=10
        )

fig.suptitle('Cargo mensual promedio segÃºn evasiÃ³n, tipo de contrato y gÃ©nero', fontsize=16)
plt.tight_layout()
plt.show()
```
